```python
# Load libraries

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# functions

def initiate_board():
    return np.full((6,6), 0, dtype=str)

def print_board(board):
    rownames = ['A', 'B', 'C', 'D', 'E', 'F']
    print(f'    1  2  3  4  5  6')
    for i in range(6):
        print(f'{rownames[i]}:  {board[i][0]+'  '+board[i][1]+'  '+board[i][2]+'  '+board[i][3]+'  '+board[i][4]+'  '+board[i][5]}')
    return

def convert_row(row):
    return ord(row)-ord('A')

def check_eligible(board, row, col):
    valid = True
    if row > 5 or col > 5:
        valid = False
    elif board[row, col] != '0':
        valid = False
    return valid

def switch_turns(player_turn):
    if player_turn=='r':
        return 'b'
    else:
        return 'r'

def push_tiles(board, row, col):
    if not row == 0:
        if board[row-1, col]!='0':
            board = push_u(board, row, col)
        if col != 0:
            if board[row-1, col-1]!='0':
                board = push_ul(board, row, col)
        if not col == 5:
            if board[row-1, col+1]!='0':
                board = push_ur(board, row, col)
    if not row == 5:
        if board[row+1, col]!='0':
            board = push_d(board, row, col)
        if not col == 0:
            if board[row+1, col-1]!='0':
                board = push_dl(board, row, col)
        if not col == 5:
            if board[row+1, col+1]!='0':
                board = push_dr(board, row, col)
    if not col == 0:
        if board[row, col-1]!='0':
            board = push_l(board, row, col)
    if not col == 5:
        if board[row, col+1]!='0':
            board = push_r(board, row, col)
    return board

def push_u(board, row, col):
    if row==1:
        board[row-1, col] = '0'
        return board
    if board[row-2, col] == '0':
        board[row-2, col] = board[row-1, col] 
        board[row-1, col] = '0'
    return board

def push_d(board, row, col):
    if row==4:
        board[row+1, col] = '0'
        return board
    if board[row+2, col] == '0':
        board[row+2, col] = board[row+1, col]
        board[row+1, col] = '0'
    return board

def push_l(board, row, col):
    if col==1:
        board[row, col-1] = '0'
        return board
    if board[row, col-2] == '0':
        board[row, col-2] = board[row, col-1]
        board[row, col-1] = '0'
    return board

def push_r(board, row, col):
    if col==4:
        board[row, col+1] = '0'
        return board
    if board[row, col+2] == '0':
        board[row, col+2] = board[row, col+1]
        board[row, col+1] = '0'
    return board

def push_ul(board, row, col):
    if row==1 or col==1:
        board[row-1, col-1] = '0'
        return board
    if board[row-2, col-2] == '0':
        board[row-2, col-2] = board[row-1, col-1]
        board[row-1, col-1] = '0'
    return board

def push_ur(board, row, col):
    if row==1 or col==4:
        board[row-1, col+1] = '0'
        return board
    elif board[row-2, col+2] == '0':
        board[row-2, col+2] = board[row-1, col+1]
        board[row-1, col+1] = '0'
    return board

def push_dl(board, row, col):
    if row==4 or col==1:
        board[row+1, col-1] = '0'
        return board
    elif board[row+2, col-2] == '0':
        board[row+2, col-2] = board[row+1, col-1]
        board[row+1, col-1] = '0'
    return board

def push_dr(board, row, col):
    if row==4 or col==4:
        board[row+1, col+1] = '0'
        return board
    if board[row+2, col+2] == '0':
        board[row+2, col+2] = board[row+1, col+1]
        board[row+1, col+1] = '0'
    return board

def print_litter(player_turn, board):
    max_litter = 8
    placed = count_kittens(player_turn, board)
    print(f'Player {player_turn} has {max_litter - placed} kittens in his litter')
    return 

def count_kittens(player_turn, board):
    return np.count_nonzero(board==player_turn)

def three_in_a_row(player_turn, board):
    for r in range(6):
        for c in range(6):
            if board[r, c] == player_turn:
                if not r > 4:
                    if check_line_down(player_turn, board, r, c):
                        return True
                    if not c > 4:
                        if check_line_diag(player_turn, board, r, c):
                            return True
                if not c > 4:    
                    if check_line_right(player_turn, board, r, c):
                        return True
                    if not r < 2:
                        if check_line_other_diag(player_turn, board, r, c):
                            return True
    return False

def check_line_right(player_turn, board, row, col):
    if board[row, col+1] == player_turn:
        if board[row, col+2] == player_turn:
            return True
    return False

def check_line_down(player_turn, board, row, col):
    if board[row+1, col] == player_turn:
        if board[row+2, col] == player_turn:
            return True
    return False

def check_line_diag(player_turn, board, row, col):
    if board[row+1, col+1] == player_turn:
        if board[row+2, col+2] == player_turn:
            return True
    return False

def check_line_other_diag(player_turn, board, row, col):
    if board[row-1, col+1] == player_turn:
        if board[row-2, col+2] == player_turn:
            return True
    return False

def check_winner(board):
    if count_kittens('r', board)==8:
        return 'r'
    if count_kittens('b', board)==8:
        return 'b'
    if three_in_a_row('r', board):
        return 'r'
    if three_in_a_row('b', board):
        return 'b'
    return 'none'
```

```python

finished = False
player_turn = 'r'
board = initiate_board()
print_board(board)

while not finished:
        # player turns
    if player_turn == 'r':
        print_litter(player_turn, board)
        row, col = input(f"Player {player_turn}: What tile would you like to place your kitten?")
        row = convert_row(row)
        col = int(col)-1
        if not check_eligible(board, row, col):
            print("Invalid input")
            continue
    elif player_turn == 'b':
        print_litter(player_turn, board)
        row, col = input(f"Player {player_turn}: What tile would you like to place your kitten?")
        row = convert_row(row)
        col = int(col)-1
        if not check_eligible(board, row, col):
            print("Invalid input")
            continue
    
    # place the kitten and push other kittens
    board[row, col] = player_turn
    board = push_tiles(board, row, col)

    print_board(board)

    # check for a winner
    winner = check_winner(board)
    if not winner == 'none':
        finished = True
        break

    player_turn = switch_turns(player_turn)

    

    

print(f'Player {winner} has won the game!')
```
